```python
# Import the required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'iris_data' is a Pandas DataFrame, you can create a scatter plot as follows:
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=iris_data) # Changed column names to match DataFrame
plt.show()
```

```python
x = iris_data.drop("Species", axis=1)
x
```

```python
y = iris_data["Species"]
y
```

```python
from sklearn.model_selection import train_test_split # Import the required module
import pandas as pd # Import pandas for DataFrame manipulation
from sklearn.neighbors import KNeighborsClassifier # Import KNeighborsClassifier

# Assuming 'x' and 'y' are your features and target variable
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42) # Split data into training and testing sets

knn = KNeighborsClassifier(n_neighbors=3) # Create a KNN classifier object
knn.fit(x_train, y_train) # Fit the classifier on the training data
```

```python
y_pred = knn.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
